# Аппроксимация функций с помощью Tensorflow

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

Генерация test, train, valid dataset

```python
x_train = np.linspace(-2 * np.pi, 2 * np.pi, 1000).astype(np.float32)
y_train = np.sin(x_train).astype(np.float32)

x_test = np.linspace(-2 * np.pi, 2 * np.pi, 300).astype(np.float32)
y_test = np.sin(x_test).astype(np.float32)
```

Базовая нейронная сеть

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(16, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(1)
])
```

```python
loss = tf.keras.losses.MeanSquaredError()
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
```

```python
model.compile(optimizer=optimizer, loss=loss)
model.fit(x_train, y_train, epochs=200, verbose=0)
```

```python
predictions = model(x_test)
```

```python
plt.plot(x_test, y_test, 'b', label='True')
plt.plot(x_test, predictions, 'r', label='Predicted')
plt.show()
```
